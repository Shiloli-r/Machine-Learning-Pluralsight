The predict function is based off the line equation: y = x * w + b (b is y-intercept and w is slope)

```python
def predict(X, w, b):
    return X * w + b
```

```python
predict(14, 1.2, 12)
```

```python
# Broadcasting input values
X = np.array([14, 5, 7])
predict(X, 1.2, 12)
```

Function to calculate the loss (how wrong the line is at approximating the dataset). The mean squared error is the loss

```python
def loss(X, Y, w, b):
    predictions = predict(X, w, b)
    return np.average((predictions - Y) ** 2)
```

```python
# Reload the dataset
X, Y = np.loadtxt("pizza.txt", skiprows=1, unpack=True)
loss(X, Y, 1.2, 12)
```

```python
def train(X, Y, iterations, lr):
    w = b = 0
    for i in range(iterations):
        current_loss = loss(X, Y, w, b)
        print("Iteration %4d => Loss: %.6f" % (i, current_loss))

        if loss(X, Y, w - lr, b) < current_loss:
            w -= lr
        elif loss(X, Y, w + lr, b) < current_loss:
            w += lr
        elif loss(X, Y, w, b - lr) < current_loss:
            b -= lr
        elif loss(X, Y, w, b + lr) < current_loss:
            b += lr
        else:
            return w, b

    raise Exception("Couldn't find a result within %d iterations" % iterations)
```

```python
w, b = train(X, Y, 10000, 0.01)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

# compute losses for w ranging from -1 to 4
weights = np.linspace(-1.0, 4.0, 200)
losses = [loss(X, Y, w, 0) for w in weights]  # b is set at a constant(0) for simplicity

# plot weights and losses
plt.axis([-1, 4, 0, 1000])
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Weight")
plt.ylabel("Loss")
plt.plot(weights, losses, color="black")

plt.show()
```

```python
def gradient(X, Y, w, b):
    """
    calculates the gradient of w and b using the partial derivatives of w and b with respect to Loss
    """
    w_gradient = np.average(2 * X * (predict(X, w, b) - Y))
    b_gradient = np.average(2 * (predict(X, w, b) - Y))
    return (w_gradient, b_gradient)
    
    
```

```python
# getting gradient of a random point in the dataset
gradient(X, Y, 2, 3)
```

Redoing the train function using gradient descent

```python
def train(X, Y, iterations, lr):
    w = b = 0
    for i in range(iterations):
        print("Iteration %4d => Loss: %.6f" % (i, loss(X, Y, w, b)))
        
        # calculate the gradient components
        w_gradient, b_gradient = gradient(X, Y, w, b)
        
        # invert the gradient, scale (multiply by a small constant -- lr) then take a step
        w -= w_gradient * lr
        b -= b_gradient * lr
    return(w, b)
         
        
```

```python
w, b = train(X, Y, 10000, 0.001)
```
