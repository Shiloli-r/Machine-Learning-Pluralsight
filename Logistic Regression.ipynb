# Multiple Regression (Logistic)

consider W0X0 + W1X1 + W2X2 + W3X3 + .... WnXn as the multiple regression equation
The input values should be in matrix form for multiple regression

Enforcing matrix dimesions

```python
import numpy as np
x1, x2, x3, y = np.loadtxt("pizza_multiple.txt", skiprows=1, unpack=True)
X = np.column_stack((np.ones(x1.size), x1, x2, x3)) # merge the 3 columns into the array (add the bias column at beginning)
Y = y.reshape(-1, 1) # reshape Y so that it has 1 colum and as many columns as necessary (-1)

```

```python
def sigmoid(z):
    return 1/ (1 + np.exp(-z))
```

```python
def predict(X, w):
    return sigmoid(np.matmul(X, w))   # matrix multiplicaiton
```

```python
# The Log Loss
```

```python
def loss(X, Y, w):
    predictions = predict(X, w)
    first_term = Y * np.log(predictions)
    second_term = (1 - Y) *  np.log(1 - predictions)
    return -np.average(first_term + second_term)
```

```python
def gradient(X, Y, w):
    return np.matmul(X.T, (predict(X, w) - Y)) / X.shape[0]
        
```

```python
def train(X, Y, iterations, lr):
    w= np.zeros((X.shape[1], 1))  # reshaping W for matrix math
    for i in range(iterations):
        print("Iteration %4d => Loss: %.6f" % (i, loss(X, Y, w)))
        w -= gradient(X, Y, w) * lr
    return w
         
```

Running Multiple Regression

```python
w = train(X, Y, iterations=10000, lr=0.001)
```

Testing predictions with original dataset

```python
predict(X[5], w)
```
